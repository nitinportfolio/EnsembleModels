# Creating features

```python
data.index.hour
```

```python
def create_features(data):
    """
    Create time series features based on time series index.
    """
    data = data.copy()
    data["hour"] = data.index.hour
    data["dayofweek"] = data.index.day_of_week
    data["quarter"] = data.index.quarter
    data["month"] = data.index.month
    data["year"] = data.index.year
    data["dayofyear"] = data.index.dayofyear
    # Monday - 0 & Sunday - 6
    return data
```

```python
data = create_features(data)
```

```python
data
```

# Visualize our Feature to Target RelationSHip

```python
fig, ax = plt.subplots(figsize = (10,8))
sns.boxplot(data = data, x = "hour", y = "PJME_MW")
ax.set_title("MW by Hour")
```

```python
fig, ax = plt.subplots(figsize = (10,8))
sns.boxplot(data = data, x = "month", y = "PJME_MW", palette = "Blues")
ax.set_title("MW by Month")
```

```python
# Above we can see Peaks around summers when everyone is running ACs
# Also during winter seasons - running heating
```

```python
train = create_features(train)
test = create_features(test)
```

```python
data.columns
```

```python
FEATURES = ['hour', 'dayofweek', 'quarter', 'month', 'year','dayofyear']
TARGET = ["PJME_MW"]
```

```python
X_train = train[FEATURES]
y_train = train[TARGET]

X_test = test[FEATURES]
y_test = test[TARGET]
```

# Creating a Regressor Model

```python
reg = xgb.XGBRegressor(n_estimators = 1000, early_stopping_rounds = 50,
                      learning_rate = 0.001)
reg.fit(X_train, y_train,
       eval_set = [(X_train, y_train),(X_test, y_test)],
       verbose = 100)
```

# Feature Importance

```python
reg.feature_importances_
```

```python
fi = pd.DataFrame(data = reg.feature_importances_,
            index = reg.feature_names_in_, 
            columns=["Importance"])
```

```python
fi.sort_values("Importance", ascending=False)
```

```python
fi.sort_values("Importance").plot(kind = "barh", title = " Feature Importance")
plt.show()
```

# Forecast on Test

```python
reg.predict(X_test)
```

```python
test["predictions"] = reg.predict(X_test)
```

```python
data_new = data.merge(test[["predictions"]], how = "left", left_index=True, right_index=True)
```

```python
ax = data_new[["PJME_MW"]].plot(figsize=(15,5))
data_new["predictions"].plot(ax = ax, style = ".")
plt.legend(["Truth","Predictions"])
ax.set_title("Raw Data & Predictions")
```

```python
data_new.loc[(data_new.index > "01-01-2018")&(data_new.index < "01-08-2018")].plot(figsize = (15,5),title =  "Week of Data")
```

```python
ax = data_new.loc[(data_new.index > "01-01-2018")&(data_new.index < "01-08-2018")]["PJME_MW"] \
.plot(figsize = (15,5),title =  "Week of Data")
data_new.loc[(data_new.index > "01-01-2018")&(data_new.index < "01-08-2018")]["predictions"] \
.plot(style = ".")
plt.legend(["Truth Data", "Predictions"])
plt.show()
```

```python
score = np.sqrt(mean_squared_error(test["PJME_MW"], test["predictions"]))
print(f'RMSE Score on Test set: {score:0.2f}')
# 2 decimal points
```

# Calculate Error

```python
test['error'] = np.abs(test["PJME_MW"] - test['predictions'])
```

```python
test
```

```python
# Worst prediction days
test['date'] = test.index.date
test.groupby(['date'])['error'].mean().sort_values(ascending=False).head(10)
```

```python
# Best prediction days
test.groupby(['date'])['error'].mean().sort_values(ascending=True).head(10)
```
