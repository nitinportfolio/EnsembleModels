```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
```

```python
np.random.seed(42)
X = np.random.rand(100,1) -0.5
y = 3*X[:,0]**2 +0.05*np.random.randn(100)

# added some randomness using 0.05*np.random.randn(100)
```

```python
plt.scatter(X,3*X[:,0]**2)
```

```python
plt.scatter(X,y)
# With Randomness
```

```python
df = pd.DataFrame()
df["X"] = X.reshape(100)
df["y"] = y
```
